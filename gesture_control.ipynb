```python
import cv2
import time
import keyboard
import pyautogui
import numpy as np
from pynput.keyboard import Controller, Key
```

```python
BLUE_MAX = np.array([128,255,255], np.uint8)
BLUE_MIN = np.array([88,78,20], np.uint8)
centroid_x = 0
centroid_y = 0
s = ''
move = ''
keyboard = Controller()
a = 0
b = 0
c = 0
d = 0
```

```python
cap = cv2.VideoCapture(0)

current_key_pressed = set()
while (cap.isOpened()):
    keyPressed = False
    key_count = 0
    key_pressed = 0
    
    ret, img = cap.read()
    img = cv2.flip(img, 1)
    img = cv2.GaussianBlur(img, (5, 5), 0)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    mask = cv2.inRange(hsv, BLUE_MIN, BLUE_MAX)
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations=2)
    
    contours, hierarchy = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if ret == True:
        if len(contours) > 0:
            contour = max(contours, key = cv2.contourArea)
            x,y,w,h = cv2.boundingRect(contour)
            cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
            centroid_x = int((x + x + w) / 2)
            centroid_y = int((y + y + h) /2)

            cv2.namedWindow("output", cv2.WINDOW_NORMAL)
            img = cv2.resize(img, (900, 700))
            
#             cv2.line(img, (0, 700), (900, 0), (255, 0, 0), 5)
#             cv2.line(img, (0, 0), (900, 700), (255, 0, 0), 5)

            cv2.line(img, (450, 0), (450, 700), (255, 0, 0), 5)
            cv2.line(img, (0, 350), (900, 350), (255, 0, 0), 5)
            cv2.imshow('Frame', img)
            cv2.getWindowImageRect('Frame')

            if centroid_x >= 0 and centroid_x <= 450:
                if centroid_y >= 0 and centroid_y <= 350:
                    a = 1
                    keyboard.press(Key.up)
                    key_pressed = Key.up
                    keyPressed = True
                    key_count = key_count + 1
                    current_key_pressed.add(Key.up)
#                 else:
#                     if a == 1:
#                         keyboard.release(Key.up)
#                         a = 0
                    
                if centroid_y >= 350 and centroid_y <=700:
                    b = 1
                    keyboard.press(Key.down)
                    key_pressed = Key.down
                    keyPressed = True
                    key_count = key_count + 1
                    current_key_pressed.add(Key.down)
                    
#                 else:
#                     if b == 1:
#                         keyboard.release(Key.down)
#                         b = 0
                        

            if centroid_x >= 451 and centroid_x <= 900:
                if centroid_y >= 0 and centroid_y <= 350:
                    c = 0
                    keyboard.press(Key.left)
                    key_pressed = Key.left
                    keyPressed = True
                    key_count = key_count + 1
                    current_key_pressed.add(Key.left)
                    
#                 else:
#                     if c == 1:
#                         keyboard.release(Key.left)
#                         c = 0
                        
                if centroid_y > 350 and centroid_y <=700:
                    d = 0
#                     keyboard.press(Key.right)
#                     key_pressed = Key.right
#                     keyPressed = True
#                     key_count = key_count + 1
#                     current_key_pressed.add(Key.right)
                
                    
                    keyPressed = False
                    key_count = 0
                    current_key_pressed = set()
#                 else:
#                     if d == 1:
#                         keyboard.release(Key.right)
#                         d = 0
        if len(current_key_pressed) > 0:
            for key in current_key_pressed:
                keyboard.release(key)
            current_key_pressed = set()

#         elif key_count > 0 and len(current_key_pressed) > 1:
#             for key in current_key_pressed:
# #                 if key_pressed != key:
#                 keyboard.release(key)
#             current_key_pressed = set()
                
          
                                     
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else: 
        break
        
cap.release()
 
# Closes all the frames
cv2.destroyAllWindows()
```
